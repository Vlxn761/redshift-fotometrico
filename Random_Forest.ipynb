```python
import tensorflow_decision_forests as tfdf
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
path_archivo ='/content/drive/MyDrive/Tareas Módulo/Proyecto redshift fotométrico/Redshift con sigma final.csv'
data = pd.read_csv(path_archivo, skiprows=[0])
```

```python
datos = data[['redshift', 'zerr', 'u', 'g', 'r', 'i', 'z']]
```

```python
datos.T
```

```python
training, validacion = np.split(datos, [int(.8*len(datos))])
train_df, test_df = np.split(training, [int(.75*len(training))])
```

```python
train_df, test_df = train_df.T, test_df.T
```

```python
# Convert the dataset into a TensorFlow dataset.
train_ds = tfdf.keras.pd_dataframe_to_tf_dataset(train_df, label="redshift", task=core.Task.REGRESSION)
test_ds = tfdf.keras.pd_dataframe_to_tf_dataset(test_df, label="redshift", task='regression')

# Train a Random Forest model.
model = tfdf.keras.RandomForestModel()
model.fit(train_ds)

# Summary of the model structure.
model.summary()

# Evaluate the model.
model.evaluate(test_ds)

# Export the model to a SavedModel
model.save("/content/drive/MyDrive/Tareas Módulo/Proyecto redshift fotométrico/ ModeloRF.keras")
```
